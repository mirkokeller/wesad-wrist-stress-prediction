# Features Notebook

## Feature Summary (48 features total)

Following the WESAD paper Section 4.1, we extract features from the wrist signals:

### ACC Features (24 features)
- Per-axis statistics (mean, std, min, max, range) × 3 axes = 15
- Per-axis peak frequency = 3
- Magnitude statistics (mean, std, min, max, range) = 5
- Magnitude peak frequency = 1

### BVP Features (9 features)
- HR: mean, std = 2
- HRV: mean NN, std NN, RMSSD = 3
- Frequency-domain energy: ULF, LF, HF, UHF = 4

### EDA Features (10 features)
- Raw signal: mean, std, min, max, range = 5
- Tonic (SCL): mean, std = 2
- Phasic (SCR): mean, std = 2
- SCR count = 1

### TEMP Features (5 features)
- Mean, std, min, max, range = 5

---

## Feature Statistics

```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("..")

import numpy as np
import pandas as pd
from collections import Counter

data = np.load('../data/03_processed/features.npz', allow_pickle=True)
X = data['X']
y = data['y']
subject = data['subject']
feature_names = list(data['feature_names'])
```

```python
print("=== DATASET SUMMARY ===\n")
print(f"Total samples: {X.shape[0]:,}")
print(f"Feature dimensions: {X.shape[1]}")
print(f"Subjects: {len(set(subject))}")
print()

print("Feature names (48):")
for i, name in enumerate(feature_names, 1):
    print(f"  {i:2d}. {name}")
```

```python
print("=== FEATURE STATISTICS ===\n")

# Organize by modality
modality_features = {
    'ACC': [f for f in feature_names if f.startswith('acc_')],
    'BVP': [f for f in feature_names if f.startswith('bvp_')],
    'EDA': [f for f in feature_names if f.startswith('eda_')],
    'TEMP': [f for f in feature_names if f.startswith('temp_')],
}

print(f"ACC features: {len(modality_features['ACC'])}")
print(f"BVP features: {len(modality_features['BVP'])}")
print(f"EDA features: {len(modality_features['EDA'])}")
print(f"TEMP features: {len(modality_features['TEMP'])}")
print(f"Total: {len(feature_names)}")
print()

# Per-modality statistics
for mod, names in modality_features.items():
    idx = [feature_names.index(n) for n in names]
    X_mod = X[:, idx]
    print(f"{mod} features:")
    print(f"  Min: {X_mod.min():.4f}, Max: {X_mod.max():.4f}")
    print(f"  Mean: {X_mod.mean():.4f}, Std: {X_mod.std():.4f}")
    nan_count = np.sum(np.isnan(X_mod))
    print(f"  NaN: {nan_count}")
    print()
```

```python
print("=== LABEL DISTRIBUTION ===\n")
label_dist = Counter(y)
total = len(y)

for lbl in sorted(label_dist.keys()):
    count = label_dist[lbl]
    pct = 100 * count / total
    print(f"Label {lbl} ({'baseline' if lbl == 1 else 'stress' if lbl == 2 else 'amusement'}): {count:,} ({pct:.1f}%)")

print(f"\nTotal: {total:,}")
```

```python
# Detailed breakdown by modality
print("=== DETAILED FEATURE BREAKDOWN ===\n")

# ACC features
print("\n--- ACC Features (24) ---")
acc_idx = [i for i, f in enumerate(feature_names) if f.startswith('acc_')]
for i in sorted(acc_idx):
    print(f"  {i:2d}. {feature_names[i]}")

# BVP features
print("\n--- BVP Features (9) ---")
bvp_idx = [i for i, f in enumerate(feature_names) if f.startswith('bvp_')]
for i in sorted(bvp_idx):
    print(f"  {i:2d}. {feature_names[i]}")

# EDA features
print("\n--- EDA Features (10) ---")
eda_idx = [i for i, f in enumerate(feature_names) if f.startswith('eda_')]
for i in sorted(eda_idx):
    print(f"  {i:2d}. {feature_names[i]}")

# TEMP features
print("\n--- TEMP Features (5) ---")
temp_idx = [i for i, f in enumerate(feature_names) if f.startswith('temp_')]
for i in sorted(temp_idx):
    print(f"  {i:2d}. {feature_names[i]}")
```

```python
# Check if subjects are balanced
from collections import Counter

subject_counts = Counter(subject)
print("=== SAMPLES PER SUBJECT ===")
for s in sorted(subject_counts.keys()):
    print(f"  S{s}: {subject_counts[s]:,}")
print(f"\nMin: {min(subject_counts.values()):,}")
print(f"Max: {max(subject_counts.values()):,}")
print(f"Mean: {np.mean(list(subject_counts.values())):.0f}")
```
